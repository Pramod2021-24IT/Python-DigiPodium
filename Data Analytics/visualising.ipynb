```python
t = np.arange(-3 * np.pi, 3 * np.pi, 0.2)
x = np.sin(t)
y = np.cos(t)
fig, ax = plt.subplots(nrows=1, ncols = 2, figsize=(15,6))
ax[0].plot(t, x, 'r')
ax[1].bar(t, y)
plt.show()
```
