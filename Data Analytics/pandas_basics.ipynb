```python
month_wise_cases = df.groupby(df.Date.dt.month)['Category'].count().reset_index()
month_wise_cases.columns = ['Month',"Case Count"]
month_wise_cases
```

```python
def num_to_months(num):
    if num == 1:
        return 'January'
    elif num == 2:
        return 'February'
    elif num == 3:
        return 'March'
    elif num == 4:
        return 'April'
    elif num == 5:
        return 'May'
    elif num == 6:
        return 'June'
    elif num == 7:
        return 'July'
    elif num == 8:
        return 'August'
    elif num == 9:
        return 'September'
    elif num == 10:
        return 'October'
    elif num == 11:
        return 'November'
    else:
        return 'December'
```

```python
#def num_to_months(num):
#    months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
#    return months[num]
```

```python
month_wise_cases['Month'] = month_wise_cases['Month'].apply(num_to_months)
```

```python
px.bar(month_wise_cases, 'Month', 'Case Count')
```
