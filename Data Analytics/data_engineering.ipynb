```python
df.dropna(subset=['price'], inplace=True)
```

```python
df.isnull().sum()   # count number of missing data in each column
```

```python
df['normalized-losses']
```

```python
df['normalized-losses'].astype('float')
```

```python
mnl = df['normalized-losses'].astype('float').mean()
```

```python
df['normalized-losses'].replace(np.nan, mnl, inplace=True)
df.head()
```

```python
df.info()
```

```python
df['normalized-losses'].replace(np.nan, mnl, inplace=True)
df['bore'].replace(np.nan, df['bore'].astype(float).mean(), inplace=True)
df['stroke'].replace(np.nan, df['stroke'].astype(float).mean(), inplace=True)
df['horsepower'].replace(np.nan, df['horsepower'].astype(float).mean(), inplace=True)
df['peak-rpm'].replace(np.nan, df['peak-rpm'].astype(float).mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df['num-of-doors'].replace(np.nan, 'four', inplace= True)
```

```python
df.isnull().sum()
```

```python
df.to_csv('automobile_cleared.csv', index= False)
```
